# Data Cleaning & Pre Processing

Remove -->  <br>'Office Space for Sale'<br>'Showroom for Sale '<br>'Shop for Sale'

```python
house_df = house_df[~house_df['Name'].str.contains(r'^(Office Space|Showroom|Shop)', case=False)]
house_df['Name'].value_counts()
```

<b>Convert--><br>'3 BHK Apartment for Sale in Shyam Vihar ' ---->['BHK : 3','Apartment','Shyam Vihar']<br>
'4 BHK House for Sale in Bangur Avenue' -----> ['BHK : 4','House','Bangur Avenue']


```python
house_df['BHK'] = house_df['Name'].str.extract(r'(\d+)\s*BHK').astype(int)
house_df['House_type'] = house_df['Name'].str.extract(r'BHK\s+(\w+)\s+for', flags=re.IGNORECASE)
house_df['Location'] = house_df['Name'].str.extract(r'in\s+(.+)', flags=re.IGNORECASE)
house_df.head()
```

<b>Area :  '1850 sqft'	(object) ---> 1850 (int)<br>

```python
house_df['Area(sqft)'] = house_df['Area'].apply(lambda x: int(x.split()[0]))
house_df.head()
```

<b>Price : <br> '₹2.38 Cr'	(object) ---> 2.38 (int) in cr <br>
 '₹32 Lac'	(object) ---> 0.32 (int) in cr <br>

```python
house_df['price(cr)'] = house_df['Price'].apply(lambda x: round(float(x.strip('₹').split()[0]) / 100, 2) if 'Lac' in x else float(x.strip('₹').split()[0]))
house_df.head()
```

<b>Drop unnecessary column

```python
house_df = house_df.drop(columns=['Price','Furnishing','Transaction','Name','Area','Status'])
house_df.head()
```

```python
house_df = house_df[['Location','House_type','BHK','Area(sqft)', 'price(cr)']]
```

# Split the dataset

```python
X = house_df.iloc[:,:-1]
y = house_df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.shape,X_test.shape
```

# Train and Test the model

```python
column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['Location', 'House_type']),
    remainder='passthrough'
)
```

```python
scaler = StandardScaler()
```

<b><h3>Linear Regression

```python
lr = LinearRegression()
pipe_lr = make_pipeline(column_trans,scaler,lr)
pipe_lr.fit(X_train,y_train)
```

```python
y_pred_lr = pipe_lr.predict(X_test)
```

<b><h3>Lasso Regression

```python
lasso = Lasso()
pipe_ls = make_pipeline(column_trans,scaler,lasso)
pipe_ls.fit(X_train,y_train)
```

```python
y_pred_lasso = pipe_ls.predict(X_test)
```

<b><h3>Ridge Regression

```python
ridge = Ridge()
pipe_rd = make_pipeline(column_trans,scaler,ridge)
pipe_rd.fit(X_train,y_train)
```

```python
y_pred_ridge = pipe_rd.predict(X_test)
```

<b><h3>R2 score :

```python
print("R2 score of Linear Regression",r2_score(y_test,y_pred_lr))
print("R2 score of Ridge Regression",r2_score(y_test,y_pred_ridge))
print("R2 score of Lasso Regression",r2_score(y_test,y_pred_lasso))
```
